# 전처리 

```python
# 결측치 확인
train.isnull().sum()
```

결측치가 겹치는 row가 많아 결측치가 많은 row 삭제
병원 종류에 따라 데이터들이 차이가 클 것 같아서 병원 종류가 없는 row 삭제
bedCount는 결측치에 같은 종류의 병원 평균을 대입
ownerChange와 employee 결측치도 대부분 겹쳐 삭제

```python
train[train['revenue1'].isnull()]
```

```python
train = train.dropna(subset=['revenue1','instkind', 'employee1', 'employee2','ownerChange'])
train.head(2)
```

```python
train.fillna(train.mean())
test.fillna(test.mean())
```

```python
train.isnull().sum().sum()
```

```python
train.shape
```

### Dtype 변경

```python
train["openDate"] = pd.to_datetime(train["openDate"], format='%Y%m%d')
train["openDate"] = train["openDate"].dt.year
```

```python
test["openDate"] = pd.to_datetime(test['openDate'], format='%Y%m%d')
test["openDate"] = test["openDate"].dt.year
```

최빈값이 80 넘어가는 column log로 변환

```python
for col in list(train.columns):
    if train[col].value_counts().max() > 80:
        print(col, train[col].dtype)
        if train[col].dtype != 'object':
            train[col] = np.log1p(train[col]+abs(train[col].min())+0.1)
            test[col] = np.log1p(test[col]+abs(test[col].min())+0.1)
```

```python
train.isnull().sum().sum()
```

## OHE

```python
train = pd.get_dummies(columns = ['instkind','sido','ownerChange'], data = train)
test = pd.get_dummies(columns = ['instkind','sido','ownerChange'], data = test)
```

```python
train = pd.get_dummies(columns = ['OC'], data = train, drop_first=True)
test = pd.get_dummies(columns = ['OC'], data = test, drop_first=True)
```

# 학습 예측 데이터셋 만들기

```python
feature_names = set(train.columns) - set(['OC_open',
                                          'sido_gangwon', 
                                          'instkind_dental_clinic', 
                                          'sido_gwangju'])
# feature_names
```

```python
label_name = 'OC_open'
```

```python
# train[label_name]
```

```python
X= train[feature_names]
y= train[label_name]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=85)
```

# 모델 만들기

## 랜덤포레스트

```python
random = RandomForestClassifier(random_state=42)
random
```

```python
cross_validate(random, X_train, y_train, cv=5, scoring="neg_mean_squared_error", error_score='raise')
```

```python
y_valid_predict = cross_val_predict(random, X_train, y_train, cv=5, n_jobs=-1)
y_valid_predict[:10]
```

```python
y_train[:10]
```

## 점수 확인

```python
#MAE
mean_absolute_error(y_train, y_valid_predict)
```

```python
mse = np.square(y_train - y_valid_predict).mean()
mse
```

```python
rmsle = (((np.log1p(y_train) - np.log1p(y_valid_predict)) ** 2).mean()) ** 0.5
rmsle
```

```python
y_predict = random.fit(X_train, y_train).predict(test[feature_names])
```

```python
df_submit = pd.read_csv("submission.csv")
```

```python
df_submit["value"] = y_predict
```

```python
file_name = f"submission_{rmsle:.5f}.csv"
```

```python
df_submit.to_csv(file_name, index=False)
```
